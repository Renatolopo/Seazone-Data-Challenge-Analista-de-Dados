## Desafio

#### 1. Ordene as cidades em ordem crescente de número de listings;

```python
city_listings_count = df.groupby('city_name')['hotel_id'].count().reset_index()

# Ordenando em ordem crescente de número de listings
city_listings_count_sorted = city_listings_count.sort_values(by='hotel_id')
city_listings_count_sorted.columns = ['cidade', 'número de listings']
print(city_listings_count_sorted)
```

#### 2. Ordene as cidades em ordem decrescente de metros quadrados

```python
# Agrupando por 'city_name' e calculando a média de metros quadrados
city_m2 = df.groupby('city_name')['room_surface_in_m2'].mean().reset_index()

# Ordenando em ordem decrescente de metros quadrados
city_m2_sorted = city_m2.sort_values(by='room_surface_in_m2', ascending=False)

print(city_m2_sorted)
```

#### 3. Quais cidades têm mais avaliações?

```python
city_ratings_count = df.groupby('city_name')['number_of_ratings'].sum().reset_index()

# Mostrando as 10 cidades com mais avaliações
city_ratings_count_sorted = city_ratings_count.sort_values(by='number_of_ratings', ascending=False)
top_city = city_ratings_count_sorted.head(10)

plt.figure(figsize=(12, 6))
plt.barh(top_city['city_name'], top_city['number_of_ratings'], color='skyblue',  height=0.4)
plt.xlabel('Número de Avaliações', fontsize=12)
plt.ylabel('Cidade', fontsize=12)
plt.title('Top 10 Cidades com Mais Avaliações')
plt.xlim(0, 101000)
plt.gca().invert_yaxis()  
for i, v in enumerate(top_city['number_of_ratings']):
    plt.text(v + 10, i, f'{v} Avaliações', color='black', va='center', fontsize=10)
plt.tight_layout()

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

Disparadamente Florianópolis é a cidade com maior número de avaliações com mais de 87 mil, onde o segundo colocado Anitápolis aparece somente com 2 mil avaliações, o gráfico acima mostra quais são as 10 cidades com mais avaliações.

#### 4. Quais cidades têm a maior média de avaliações? E a menor média? a. Existe alguma explicação para isso? Conjecture.

```python
media_avaliacoes_por_cidade = df.groupby('city_name')[['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício', 'Localização', 'Total', 'WiFi gratuito']].mean()

# Calculando a média geral de avaliações por cidade
media_avaliacoes_por_cidade = media_avaliacoes_por_cidade.mean(axis=1).reset_index()
media_avaliacoes_por_cidade.columns = ['city_name', 'media_avaliacoes']


top5_maior_media = media_avaliacoes_por_cidade.sort_values(by='media_avaliacoes', ascending=False).head(5)
top5_menor_media = media_avaliacoes_por_cidade.sort_values(by='media_avaliacoes', ascending=True).head(5)

df_concatenado = pd.concat([top5_maior_media, top5_menor_media])


plt.figure(figsize=(10, 6))
sns.barplot(data=df_concatenado, x='media_avaliacoes', y='city_name', palette="coolwarm")
plt.title('Top 5 Cidades com Maiores e Menores Médias de Avaliações')
plt.xlabel('Média de Avaliações')
plt.ylabel('Cidade')

for i in range(10):
    plt.text(df_concatenado['media_avaliacoes'].iloc[i], i, round(df_concatenado['media_avaliacoes'].iloc[i], 2), ha='left')

plt.show()
```

As cidades com maior/menor média de avaliação estão no gráfico acima onde fiz um top 5 para cada categoria.


```python
# tentando identificar uma explicação para o caso acima analisando as caracteristicas dos imóveis de cada cidade
top_maior_menor = pd.concat([top5_maior_media, top5_menor_media])

df_aux = df[['city_name', 'Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício', 'Localização', 'Total', 'WiFi gratuito', 'room_surface_in_m2', 'number_of_ratings']]
df_aux = df_aux[df_aux['city_name'].isin(top_maior_menor['city_name'])]
df_aux = df_aux.groupby('city_name').mean().reset_index()
df_aux
```

```python
# O Tamanho parece ser um fator importante, irei criar um boxplot para entender melhor
df_aux['Cidades com menor média'] = df_aux[df_aux['city_name'].isin(top5_menor_media['city_name'])]['room_surface_in_m2']
df_aux['Cidades com maior média']  = df_aux[df_aux['city_name'].isin(top5_maior_media['city_name'])]['room_surface_in_m2']


plt.figure(figsize=(8, 6))  
sns.set(style='whitegrid') 
boxplot = sns.boxplot(data=df_aux[['Cidades com menor média', 'Cidades com maior média']], palette='Set3')
boxplot.set(xlabel="Variáveis", ylabel="Valores")  
plt.title('Boxplot do tamanho do quarto/imóvel', fontsize=16)  
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12) 
plt.show()
```

Analisando quais variáveis podem explicar o porque algumas cidades têm nota maior que outras, descobri que o tamanho do imóvel é uma característica que pode influenciar bastante, no gráfico acima podemos concluir que no geral as cidades que tem uma média de tamanho dos imóveis menor, estão entre as cidades pior avaliadas, já as cidades que tem uma  média de tamanho de imovel maior tem uma avaliação melhor.

#### 5. Existem correlações entre as características de um anúncio e a sua localização? a. Quais? Explique

```python
# Nova coluna no df com numero de room_facilities que sera importante nessa questão e tambem será usada nas proximas questões
df['n_room_facilities'] = df['room_facilities'].apply(lambda x: len(x.split(',')))
```

```python
# proximo passo agora e tratar a base para deixar somente colunas numericas para medir a correlação,
# para isso irei remover algumas colunas como url e converter para numerica as colunas do tipo texto que serão usadas 
df2 = df.drop(columns=['hotel_id', 'url','room_id', 'hotel_name_y', 'Unnamed: 0_y'], axis=1)
df2['city_name'] = pd.Categorical(df['city_name']).codes
df2['hotel_name_x'] = pd.Categorical(df['hotel_name_x']).codes
df2['accommodation_type'] = pd.Categorical(df['accommodation_type']).codes
df2['room_name'] = pd.Categorical(df['room_name']).codes
df2['room_facilities'] = pd.Categorical(df['room_facilities']).codes
df2
```

```python
# Medindo a correlação da coluna Localização com as demais
correlation = df2.corrwith(df2['Localização'])
correlation_df = pd.DataFrame(correlation, columns=['Correlation'])

plt.figure(figsize=(10, 6))  
sns.heatmap(correlation_df, annot=True, fmt=".1f", linewidths=.6, cmap='coolwarm')
plt.title('Correlações com a coluna Localização')
plt.show()
```

Observando o gráfico acima podemos concluir que a coluna "Localização" onde temos a nota que os usuários avaliam a localização tem uma correlação moderada/fraca com as outras colunas de avaliação (Comodidades, Limpeza, Conforto, Custo-benefício e Total), Já as demais colunas não possuem correlação com a nota da localização.

#### 6. Existem relações entre a nota do anúncio e os recursos disponíveis no imóvel? a. Quais? Explique

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='n_room_facilities', y='Total', data=df)
plt.title('Relação entre Recursos disponíveis e Total')
plt.xlabel('Recursos disponíveis')
plt.ylabel('Total')
plt.show()
```

Por esse gráfico é possível identificar que as notas que estão entre 8-10 têm um número de recursos concentrados entre 20 e 40. Porém não dá para concluir que ter pouco recurso disponível a nota vai ser baixa, pois se olharmos para os imóveis com poucos recursos eles estão com grande parte das notas altas também. Próximo passo é verificar a correlação desses dados.

```python
# verificando se a quantidade de recursos tem correlação com alguma nota
df2 = df2[['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício','Total', 'n_room_facilities']]

correlation = df2.corrwith(df2['n_room_facilities'])
correlation_df = pd.DataFrame(correlation, columns=['Correlation'])

plt.figure(figsize=(10, 6))  
sns.heatmap(correlation_df, annot=True, fmt=".1f", linewidths=.6, cmap='coolwarm')
plt.title('Correlações com a coluna n_room_facilities')
plt.show()
```

O gráfico acima não dá pra concluir muita coisa ainda, pois estamos comparando somente a quantidade de recursos disponíveis, mas talvez o que pode afetar a nota são quais recursos estão disponíveis e não a quantidade. Para isso vou criar novas colunas para indicar a presença de um recurso específico no anúncio e medir sua correlação com a nota.


```python
# esses  serão os rucursos considerados nos anuncios, será criado uma nova coluna pra cada recuros que recebera 1 se o recuros estiver presente no  anuncio ou 0 caso contrario
facilities = ['Cozinha', 'Geladeira', 'Fogão', 'Varanda', 'Sofá', 'Papel higiênico', 'Ventilador', 'Mesa de jantar', 'Toalhas', 'Churrasqueira', 'Vista', 'TV', 'Micro-ondas', 'Internet', 'Varanda', 'Ar-condicionado', 'Roupa de cama', 'Utensílios de cozinha', 'Banheiro privativo', 'Extintor de incêndio','Secador de cabelo', 'Máquina de café', 'Vista do mar', 'Vaso sanitário', 'Torradeira', 'Tomada perto da cama']
df2['room_facilities'] = df['room_facilities']


for i in facilities:
    df2[i] = df2['room_facilities'].apply(lambda x: 1 if i in x else 0)

# Novas colunas criadas, agora irei medir a correlação dessas colunas com a coluna de avaliação Total

df_room_facilities = df2.drop(columns=['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício','room_facilities']) # removendo colunas que contem texto e as que não utilizarei
correlation = df_room_facilities.corrwith(df_room_facilities['Total'])
correlation_df = pd.DataFrame(correlation, columns=['Correlation'])

plt.figure(figsize=(10, 6))  
sns.heatmap(correlation_df, annot=True, fmt=".1f", linewidths=.6, cmap='coolwarm')
plt.title('Correlações com a coluna Total')
plt.show()
```

Usando a correlação entre as variáveis nenhuma correlação passou de 0.3 para possuir pelo menos uma correlação fraca, então não foi possível identificar relações entre a nota do anúncio e os recursos disponíveis no imóvel.

#### 7. Existe alguma relação entre a nota recebida e a localização? a. Quais? Explique.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Localização', y='Total', data=df)
plt.title('Relação entre Localização e Total')
plt.xlabel('Localização')
plt.ylabel('Total')
plt.show()


correlation = df['Localização'].corr(df['Total'])
mean_total = df['Total'].mean()
std_total = df['Total'].std()

print(f"A correlação entre Localização e Total é: {correlation}")
print(f"Média das notas 'Total': {mean_total}")
print(f"Desvio padrão das notas 'Total': {std_total}")
```

A correlação de 0.49 indica uma correlação positiva moderada entre a localização e as notas recebidas, ou seja, à medida que a localização melhora, as notas totais também tendem a melhorar. No entanto, é importante notar que a correlação não implica causalidade. Ou seja, não podemos afirmar que uma boa localização é a única causa das boas notas. Outros fatores também podem influenciar as avaliações, como a qualidade das comodidades, limpeza, conforto, custo-benefício, entre outros.

#### 9. Quais são os anúncios que te parecem críticos? a. Explique.


Anúncios críticos podem ser considerados aqueles que tem uma nota abaixo de 4, podemos identificar eles da seguinte forma

```python
limite_critico = 4.0

nota_media_por_anuncio = df.groupby(['hotel_id', 'hotel_name_x'])['Total'].mean().reset_index()

anuncios_criticos = nota_media_por_anuncio[nota_media_por_anuncio['Total'] < limite_critico]

print("Anúncios Críticos:")
print(anuncios_criticos)
```

Esses são alguns exemplos de anúncio crítico pois suas notas não passam de 4, que é algo bem negativo que não é visto com bons olhos pelos clientes que se deparam com esses anúncios, muitas das vezes preferem pegar um hotel sem avaliações do que com essa avaliação, pois como podemos ver até alguns casos com nota 1 indicando que esse é um péssimo imovel, dificultando que alguém possa alugar ele.
